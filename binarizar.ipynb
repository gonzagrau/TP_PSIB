```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from my_img_processor import My_Image
```

```python
img_path = r'Dataset for Fetus Framework\Dataset for Fetus Framework\External Test Set\Standard\1383.png'
fetus_img = My_Image(img_path)
plt.imshow(fetus_img, vmin=0, vmax=255, cmap='gray')
```

```python
new_img = np.zeros_like(fetus_img)
THRES = 20
for i in range(new_img.shape[0]):
    for j in range(new_img.shape[1]):
        if fetus_img[i, j] > THRES:
            new_img[i, j] = 255
```

```python
%matplotlib widget
```

```python
plt.rcParams['backend'] = 'TkAgg'
plt.ion()
x = 0
y = 0

def onclick(event):
   global x
   x = int(np.ceil(event.xdata))
   global y 
   y = int(np.ceil(event.ydata))
   print(f"{x=}, {y=}")

fig, ax = plt.subplots(figsize=(10, 10))
ax.imshow(new_img, vmin=0, vmax=255, cmap='gray')
ax.
plt.axis(False)
fig.canvas.mpl_connect('button_press_event', onclick)
plt.show()
```
