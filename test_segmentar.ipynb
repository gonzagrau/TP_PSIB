```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('N1_evoked_raw_100_F1_R1.csv')
sig_avg = df['both']
t = df['t']

plt.plot(t, sig_avg)
```

```python
T_MIN = 20 # ms
T_MAX = 40 # ms
fs = int(len(t)/t[len(t) - 1])
sp_min = int(fs*T_MIN)
sp_max = int(fs*T_MAX)
print(f"{t[sp_min]=:.2f}, {t[sp_max]=:.2f}")
print(f"{sp_min=}, {sp_max=}")
amlramp_start = np.argmax(sig_avg[sp_min: sp_max]) + sp_min
amlramp_end = np.argmin(sig_avg[sp_min: sp_max]) + sp_min
print(f"{t[amlramp_start]=:.2f}, {t[amlramp_end]=:.2f}")
print(f"{amlramp_start=}, {amlramp_end=}")


plt.plot(t, sig_avg, label='PE')
plt.plot(t[amlramp_start: amlramp_end], sig_avg[amlramp_start: amlramp_end], label='amlramp')
plt.xlabel('t [ms]')
plt.ylabel('v [nv]')
plt.title('Rampa aml segmentada')
plt.legend()
plt.show()
```
